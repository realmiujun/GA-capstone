# *First blood model*

```python
import pandas as pd
from sklearn.metrics import log_loss
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier,VotingClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.svm import SVC
from keras.models import Sequential
from keras.layers import Dense,Dropout
from tensorflow.keras.wrappers.scikit_learn import KerasRegressor
from keras import regularizers
from sklearn.metrics import f1_score
import numpy as np
import time
import pickle
```

```python
try:
    df = pd.read_csv('./model_data/model_data_fb.csv')
except:
    df = pd.read_csv('https://dl.dropboxusercontent.com/s/pj6wmr1pm521p9k/model_data_fb.csv?dl=0')
X = df.drop(columns='firstblood_blue')
y = df['firstblood_blue']
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=8)
```

```python
def log_model():
    pipe = Pipeline([
        ('lr',LogisticRegression(solver='liblinear'))
    ])

    param = {
        'lr__penalty':['l1','l2','elasticnet']
    }

    gs = GridSearchCV(pipe,param,cv=5)
    t0 = time.time()
    gs.fit(X_train,y_train)
    t1 = time.time()
    print(f'It took {t1-t0} seconds to fit')
    print(gs.score(X_test,y_test))
    print(f'f1 score:{f1_score(y_test,gs.best_estimator_.predict(X_test))}')
    print(f'log loss: {log_loss(y_test,gs.predict_proba(X_test))}')
    print(gs.best_params_)
    pred_proba = [i[1] for i in gs.best_estimator_.predict_proba(X_test)]
    pred_df = pd.DataFrame({'true_values': y_test,
                            'pred_probs':pred_proba})
    # Create figure.
    plt.figure(figsize = (10,7))

    # Create two histograms of observations.
    plt.hist(pred_df[pred_df['true_values'] == 0]['pred_probs'],
             bins = 25,
             color = 'r',
             alpha = 0.6,
             label = 'Outcome = Red')
    plt.hist(pred_df[pred_df['true_values'] == 1]['pred_probs'],
             bins = 25,
             color = 'b',
             alpha = 0.6,
             label = 'Outcome = Blue')
    plt.vlines(x=.5,
              ymin=0,
              ymax=25,
              color='grey',
              linestyle='dashdot')

    # Label axes.
    plt.title('Distribution of P(Outcome = Blue win)', fontsize = 22,color='white')
    plt.ylabel('Frequency', fontsize = 18,color='white')
    plt.xlabel('Predicted Probability that Outcome = Blue win', fontsize = 18,color = 'white')
    plt.tick_params(axis='x', colors='white')
    plt.tick_params(axis='y', colors='white')
    # Create legend.
    plt.legend(fontsize = 20);
    #Thank you Matt for the code from our lecture!
    #plt.savefig('log_plot.png');
```

```python
log_model()
```

```python
def rf_model():
    pipe = Pipeline([
        ('rf',RandomForestClassifier())
    ])
    pipe_params = {
        'rf__n_estimators':[100],
        'rf__max_features':['auto'],
        'rf__max_depth':[20]
    }
    gs = GridSearchCV(pipe,pipe_params,cv=5)
    t0 = time.time()
    gs.fit(X_train,y_train)
    t1 = time.time()
    print(f'It took {t1-t0} seconds to fit')
    print(gs.score(X_test,y_test))
    print(f'f1 score:{f1_score(y_test,gs.best_estimator_.predict(X_test))}')
    print(f'log loss: {log_loss(y_test,[i[1] for i in gs.predict_proba(X_test)])}')
    print(gs.best_params_)
    
    filename = 'rf_model.sav'
    pickle.dump(gs.best_estimator_, open(filename, 'wb'))
    
    pred_proba = [i[1] for i in gs.best_estimator_.predict_proba(X_test)]
    pred_df = pd.DataFrame({'true_values': y_test,
                            'pred_probs':pred_proba})
    # Create figure.
    plt.figure(figsize = (10,7))

    # Create two histograms of observations.
    plt.hist(pred_df[pred_df['true_values'] == 0]['pred_probs'],
             bins = 25,
             color = 'r',
             alpha = 0.6,
             label = 'Outcome = Red')
    plt.hist(pred_df[pred_df['true_values'] == 1]['pred_probs'],
             bins = 25,
             color = 'b',
             alpha = 0.6,
             label = 'Outcome = Blue')
    plt.vlines(x=.5,
              ymin=0,
              ymax=25,
              color='grey',
              linestyle='dashdot')

    # Label axes.
    plt.title('Distribution of P(Outcome = Blue win)', fontsize = 22,color='white')
    plt.ylabel('Frequency', fontsize = 18,color='white')
    plt.xlabel('Predicted Probability that Outcome = Blue win', fontsize = 18,color = 'white')
    plt.tick_params(axis='x', colors='white')
    plt.tick_params(axis='y', colors='white')
    # Create legend.
    plt.legend(fontsize = 20);
    #Thank you Matt for the code from our lecture!
    #plt.savefig('log_plot.png');
```

```python
def model_func(layer_one_neurons=32, layer_one_dropout=.2, layer_two_neurons=32):
    # print('layer_one_neurons', layer_one_neurons)
    model = Sequential()
    model.add(Dense(layer_one_neurons, activation='relu', input_dim=X_train.shape[1],kernel_regularizer=regularizers.l2()))
    model.add(Dropout(layer_one_dropout))
    model.add(Dense(layer_two_neurons, activation='relu',kernel_regularizer=regularizers.l2()))
    model.add(Dense(1, activation=None))
    model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['acc'])
    return model
nn = KerasRegressor(build_fn=model_func, batch_size=512, verbose=0)
params = {
    'epochs': [10],
    'layer_one_neurons': [34],
    'layer_two_neurons': [26],
    'layer_one_dropout': [.4]
}
gs = GridSearchCV(nn, param_grid=params, cv=2)
gs.fit(X_train, y_train)
print(gs.best_score_)
gs.best_params_
```

```python
model = Sequential()
model.add(Dense(34,input_dim=X_train.shape[1],activation='relu',kernel_regularizer=regularizers.l2()))
model.add(Dropout(.4))
model.add(Dense(26,activation='relu',kernel_regularizer=regularizers.l2()))
model.add(Dropout(.4))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['acc'])
history = model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=20,batch_size=512,verbose=2)    
print(f'f1 score:{f1_score(y_test,np.where(model.predict(X_test)>.5,1,0))}')
plt.figure(figsize=(12,7))
plt.title('Epoch accuracy', fontsize = 22,color='white')
plt.ylabel('Accuracy', fontsize = 18,color='white')
plt.xlabel('Epochs', fontsize = 18,color = 'white')
plt.plot(history.history['acc'], label='training acc', color='b')
plt.plot(history.history['val_acc'], label='testing acc', color='r')
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')
plt.legend();
```

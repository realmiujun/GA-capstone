```python
import pandas as pd
```

```python
df = pd.read_csv('./processed_data/pregame.csv')
```

```python
df
```

```python
df['win']=df['win'].map({False:0,True:1})
```

```python
df['spells'] = df['spell1Id'] + ',' + df['spell2Id']
df['spells']
```

```python
def listit(x):
    return x.split(sep=',')
df['spells']=df['spells'].map(listit)
```

```python
for i in range(len(df['spells'])):
    df['spells'][i].sort()
```

```python
df.to_csv('./processed_data/tableau_df.csv',index=False)
```
